### Paso 5: Seleccionar Países
* Selecciona los países donde se expandirán: Colombia, Argentina, Chile, México, Perú y Brasil.

```python
# Seleccionar los países
paises_seleccionados = ['Colombia', 'Argentina', 'Chile', 'Mexico', 'Peru', 'Brazil'] # Creamos el dataframe con los paises que vamos a trabajar
df_latinoamerica_paises_filtrado = df_latinoamerica[df_latinoamerica['country_name'].isin(paises_seleccionados)] #Utilizamos isin() para filtrar lo pedido

print(f"Datos filtrados por países seleccionados. Registros: {df_latinoamerica_paises_filtrado.shape[0]}") # Imprime cantidad de registros para ver la dimencion de la data
```

### Paso 6: Filtrar por Fechas Mayores a 2021-01-01
* Filtra los datos en fechas mayores a 2021-01-01.

```python
# Filtrar los datos en fechas mayores a 2021-01-01
df_latinoamerica_paises_filtrado['date'] = pd.to_datetime(df_latinoamerica_paises_filtrado['date']) # Declaramos cual es la Columna de Fechas
df_latinoamerica_paises_filtrado_mayor_2021 = df_latinoamerica_paises_filtrado[df_latinoamerica_paises_filtrado['date'] > '2021-01-01'] # Aplicamos el filtro con la condicion pedida

print(f"Datos filtrados por fecha. Registros: {df_latinoamerica_paises_filtrado_mayor_2021.shape[0]}") # Imprime cantidad de registros para ver la dimencion de la data
```

### Paso 7: Completar Valores Faltantes a Nivel de País
* Compara a nivel de país para llenar valores faltantes.

* verificamos los nulos

```python
df_latinoamerica_paises_filtrado_mayor_2021.isnull().sum()[df_latinoamerica_paises_filtrado_mayor_2021.isnull().sum()>4000000] # Creamos una mascara y filtramos para ver que columnas tienen nulos mayores a 4 millones
```

* Despues de haber hecho un analisis de la calidad de los datos, podemos ver que tenemos datos faltantes en muchas columnas, por lo que despues realizaremos un nuevo filtro por la columna "Location_key" pero solo vamos a ustilizar las instancias que correspondan a los pafses que hemos seleccionado y no a sus regiones o subregiones AR, CL, CO, MX, PE, BR.

```python
df_latinoamerica_paises_filtrado_nuevo=df_latinoamerica_paises_filtrado_mayor_2021[df_latinoamerica_paises_filtrado_mayor_2021['location_key'].isin(['AR','CL','CO','MX','PE','BR'])] #Creamos la mascara para aplicar el nuevo filtrado por Paises sin regiones y subregiones

print(f"Datos filtrados por países nuevos. Registros: {df_latinoamerica_paises_filtrado_nuevo.shape[0]}") # Imprime cantidad de registros para ver la dimencion de la data
```

```python
#contamos valores nulos 
df_latinoamerica_valores_nulos=df_latinoamerica_paises_filtrado_nuevo.isnull().sum()[df_latinoamerica_paises_filtrado_nuevo.isnull().sum()>0] #crear mascara y guardar lo nulos que hay nuevos
print(df_latinoamerica_valores_nulos) # imprimir los nulos
print(np.shape(df_latinoamerica_paises_filtrado_nuevo)) # Imprime cantidad de registros
```

### Paso 8: Limpieza Preliminar de Datos
* Realiza una limpieza preliminar de los datos, eliminando registros nulos y corrigiendo los tipos de datos donde sea necesario, trata con valores medios, valores anteriores o valores siguientes.

```python
# Para usar el for para completar los valores nulos de las columnas de la media de cada país.
for pais in paises_seleccionados:
    df_latinoamerica_paises_filtrado_nuevo['new_confirmed'] = df_latinoamerica_paises_filtrado_nuevo['new_confirmed'].fillna(df_latinoamerica_paises_filtrado_nuevo[df_latinoamerica_paises_filtrado_nuevo['country_name'] == pais]['new_confirmed'].mean())
    df_latinoamerica_paises_filtrado_nuevo['new_deceased'] = df_latinoamerica_paises_filtrado_nuevo['new_deceased'].fillna(df_latinoamerica_paises_filtrado_nuevo[df_latinoamerica_paises_filtrado_nuevo['country_name'] == pais]['new_deceased'].mean())
    df_latinoamerica_paises_filtrado_nuevo['cumulative_confirmed'] = df_latinoamerica_paises_filtrado_nuevo['cumulative_confirmed'].fillna(df_latinoamerica_paises_filtrado_nuevo[df_latinoamerica_paises_filtrado_nuevo['country_name'] == pais]['cumulative_confirmed'].mean())
    df_latinoamerica_paises_filtrado_nuevo['cumulative_deceased'] = df_latinoamerica_paises_filtrado_nuevo['cumulative_deceased'].fillna(df_latinoamerica_paises_filtrado_nuevo[df_latinoamerica_paises_filtrado_nuevo['country_name'] == pais]['cumulative_deceased'].mean())
    df_latinoamerica_paises_filtrado_nuevo['average_temperature_celsius'] = df_latinoamerica_paises_filtrado_nuevo['average_temperature_celsius'].fillna(df_latinoamerica_paises_filtrado_nuevo[df_latinoamerica_paises_filtrado_nuevo['country_name'] == pais]['average_temperature_celsius'].mean())
    df_latinoamerica_paises_filtrado_nuevo['minimum_temperature_celsius'] = df_latinoamerica_paises_filtrado_nuevo['minimum_temperature_celsius'].fillna(df_latinoamerica_paises_filtrado_nuevo[df_latinoamerica_paises_filtrado_nuevo['country_name'] == pais]['minimum_temperature_celsius'].mean())
    df_latinoamerica_paises_filtrado_nuevo['maximum_temperature_celsius'] = df_latinoamerica_paises_filtrado_nuevo['maximum_temperature_celsius'].fillna(df_latinoamerica_paises_filtrado_nuevo[df_latinoamerica_paises_filtrado_nuevo['country_name'] == pais]['maximum_temperature_celsius'].mean())
    df_latinoamerica_paises_filtrado_nuevo['rainfall_mm'] = df_latinoamerica_paises_filtrado_nuevo['rainfall_mm'].fillna(df_latinoamerica_paises_filtrado_nuevo[df_latinoamerica_paises_filtrado_nuevo['country_name'] == pais]['rainfall_mm'].mean())
    df_latinoamerica_paises_filtrado_nuevo['relative_humidity'] = df_latinoamerica_paises_filtrado_nuevo['relative_humidity'].fillna(df_latinoamerica_paises_filtrado_nuevo[df_latinoamerica_paises_filtrado_nuevo['country_name'] == pais]['relative_humidity'].mean())

# Llenar NaN con el valor de 0 en la columna 'new_recovered', 'cumulative_recovered', 'cumulative_vaccine_doses_administered' para hacer análisis de estos datos.
df_latinoamerica_paises_filtrado_nuevo['new_recovered'] = df_latinoamerica_paises_filtrado_nuevo['new_recovered'].fillna(0)
df_latinoamerica_paises_filtrado_nuevo['cumulative_recovered'] = df_latinoamerica_paises_filtrado_nuevo['cumulative_recovered'].fillna(0)
df_latinoamerica_paises_filtrado_nuevo['cumulative_vaccine_doses_administered'] = df_latinoamerica_paises_filtrado_nuevo['cumulative_vaccine_doses_administered'].fillna(0)
```

### Paso 12: Crear Función para Mediana, Varianza y Rango
*   Crea una función que permita obtener la mediana, varianza y el rango. Te dejamos algunas preguntas para analizar esas medidas:

    ¿Qué representa la mediana?

    ¿Cómo varía la dispersión de los datos en el conjunto de datos analizado, en términos de la varianza y el rango? 

    ¿Qué nos puede indicar esto sobre la consistencia o la variabilidad de los datos en relación con la mediana?

```python
# Crear una función que solo muestre los valores numéricos y no los valores categóricos, además solo da la media, mediana y la desviación estándar.
def calcular_estadisticas(data):
    for columna in data.columns:
        if data[columna].dtype == 'float64' or data[columna].dtype == 'int64':
            print(columna)
            print('Mediana:', data[columna].median())
            print('Varianza: ', np.around((data[columna].var()),decimals=2)) # Casteo (redondeo) a 2 decimales
            print('Rango:', data[columna].max() - data[columna].min())
            print('Moda:', data[columna].mode()[0])
            print('--------------------------------')

calcular_estadisticas(data_latinoamerica_final)
```
